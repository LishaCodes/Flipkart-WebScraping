```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
def get_title(soup):
    title = soup.find("a", attrs={"class": "_31qSD5"})
    if title is not None:
        return title.text.strip()
    return ""

def get_price(soup):
    price = soup.find("div", attrs={"class": "_1_WHN1"})
    if price is not None:
        return price.text.strip()
    return ""

def get_rating(soup):
    rating = soup.find("div", attrs={"class": "_3LWZlK"})
    if rating is not None:
        return rating.text.strip()
    return ""

def get_review_count(soup):
    review_count = soup.find("span", attrs={"class": "_2_R_DZ"})
    if review_count is not None:
        return review_count.text.strip()
    return ""

def get_availability(soup):
    availability = soup.find("div", attrs={"class": "_9-sL7L"})
    if availability is not None:
        return availability.text.strip()
    return "Not Available"
```

```python
if __name__ == '__main__':
    URL = "https://www.flipkart.com/search?q=mobiles"

    webpage = requests.get(URL)
    soup = BeautifulSoup(webpage.content, "html.parser")

    products = soup.find_all("div", attrs={"class": "_1AtVbE"})

    data = []

    for product in products:
        title = get_title(product)
        price = get_price(product)
        rating = get_rating(product)
        review_count = get_review_count(product)
        availability = get_availability(product)

        data.append({
            'title': title,
            'price': price,
            'rating': rating,
            'review_count': review_count,
            'availability': availability
        })

    flipkart_df = pd.DataFrame(data)
    flipkart_df.dropna(subset=['title'], inplace=True)
    flipkart_df.to_csv("flipkart_data.csv", header=True, index=False)
```
